```python
#Test vector
col=[1,6,3,7,5,5,7,3,6,7,8,3,7,5,4,3,4,34,5,56,56]
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
# Moving mean function
def moving_mean(data,width):
    val=np.zeros(width)
    moving_mean = []
    for i in data:
        val = np.append(np.delete(val,0),i)
        meanv = np.mean(val)
        moving_mean.append(meanv)
    return(moving_mean)
```

```python
moving_mean(col,3)
```

```python
plt.plot(moving_mean(col,3))
```
